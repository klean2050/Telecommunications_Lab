```python
import numpy as np
import matplotlib.pyplot as plt

A   = 1
fm  = 9000.0    # ID 03115117
fs1 = fm * 20
fs2 = fm * 100
fs3 = fm * 5

def sample_time(f,fs):
    return np.arange(0, 4/f, 1/fs)

def sampled_signal(f,fs):
    return A * np.sin(2*np.pi*f*sample_time(f,fs))

plt.plot(sample_time(fm,fs1), sampled_signal(fm,fs1), 'o')
plt.xlabel("Time (sec)")
plt.ylabel("Amplitude (V)")
plt.title("Exercise 1ai")
plt.show()
```
